## Data Cleaning & Visualization Project

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("employee_salary_dataset.csv")

print("Missing Values:")
print(df.isnull().sum())

df = df.fillna(0)

df = df.drop_duplicates()

Q1 = df["Monthly_Salary"].quantile(0.25)
Q3 = df["Monthly_Salary"].quantile(0.75)
IQR = Q3 - Q1

df = df[(df["Monthly_Salary"] >= Q1 - 1.5 * IQR) &
        (df["Monthly_Salary"] <= Q3 + 1.5 * IQR)]

df.to_csv("cleaned_data.csv", index=False)

plt.hist(df["Monthly_Salary"])
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Employees")
plt.show()

df["Department"].value_counts().plot(kind="bar")
plt.title("Employees in Each Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()

print("Data Cleaning & Visualization Completed!")
```
